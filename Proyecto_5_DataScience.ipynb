## Agregar datos por usuario



```python
# Calcula el número de llamadas hechas por cada usuario al mes. Guarda el resultado.


# Extraer año-mes de la fecha
df_calls_clean['year_month'] = df_calls_clean['call_date'].dt.to_period('M')

# Agrupar por usuario y mes y calcular las llamadas hechas
df_calls_monthly = df_calls_clean.groupby(['user_id', 'year_month'])['duration_round'].count()
print(df_calls_monthly.head(15))

```

```python
# Calcula la cantidad de minutos usados por cada usuario al mes. Guarda el resultado.

#agrupar por usuarios y sumar los minutos:
df_calls_group = df_calls_clean.groupby(['user_id', 'year_month'])['duration_round'].sum()

print(df_calls_group.head())

```

```python
# Calcula el número de mensajes enviados por cada usuario al mes. Guarda el resultado.

print(df_messages.info())

# Extraer año-mes de la fecha
df_messages['year_month'] = df_messages['message_date'].dt.to_period('M')

# Agrupar por usuario y mes y calcular las llamadas hechas
df_messages_monthly = df_messages.groupby(['user_id', 'year_month'])['message_date'].count()
print(df_messages_monthly.head(15))
```

```python
# Calcula el volumen del tráfico de Internet usado por cada usuario al mes. Guarda el resultado.

# Extraer año-mes de la fecha
df_internet['year_month'] = df_internet['session_date'].dt.to_period('M')

# Agrupar por usuario y mes y calcular las llamadas hechas
df_internet_monthly = df_internet.groupby(['user_id', 'year_month'])['mb_used'].sum()

#Conversión de consumo de internet a Gb:
df_internet_monthly_gb = df_internet_monthly / 1024

#Redondeo del consumo al siguiente número:
df_internet_monthly_gb_round = np.ceil(df_internet_monthly_gb)

print(df_internet_monthly_gb_round.head(15))

```

<div class="alert alert-block alert-success">
<b>Comentario de la revisora Iteración #2</b> <a class="tocSkip"></a>

Excelente uso del `groupby` para los diferentes dataframes y muy buen uso de `rename` para guardar los datos calculados en una nueva columna. </div>

[Junta los datos agregados en un DataFrame para que haya un registro que represente lo que consumió un usuario único en un mes determinado.]

```python
# Fusiona los datos de llamadas, minutos, mensajes e Internet con base en user_id y month

df_user_monthly = pd.concat([df_calls_monthly.rename('calls_made'), df_calls_group.rename('minutes_used'), df_messages_monthly.rename('messages_sent'), df_internet_monthly_gb_round.rename('gb_used')], axis=1).fillna(0)

print(df_user_monthly.head(20))
```

```python
# Añade la información de la tarifa

df_user_monthly = df_user_monthly.reset_index().merge(df_users_clean[['user_id', 'plan']], on='user_id')

print(df_user_monthly.head(15))
```

Cálculo de los ingresos en los planes

```python
# Calcula el ingreso mensual para cada usuario

#Función para el cálculo del ingreso mensual
def calculate_revenue(row):
    plan_info = df_plans_clean.loc[row['plan']]
    
    # Pago base mensual
    revenue = plan_info['usd_monthly_pay']
    
    # Cargos adicionales por minutos
    extra_minutes = max(0, row['minutes_used'] - plan_info['minutes_included'])
    revenue += extra_minutes * plan_info['usd_per_minute']
    
    # Cargos adicionales por mensajes
    extra_messages = max(0, row['messages_sent'] - plan_info['messages_included'])
    revenue += extra_messages * plan_info['usd_per_message']
    
    # Cargos adicionales por GB
    extra_gb = max(0, row['gb_used'] - plan_info['gb_per_month_included'])
    revenue += extra_gb * plan_info['usd_per_gb']
    
    return revenue

# Aplicar la función calculate_revenue a cada fila del DataFrame df_user_monthly:
df_user_monthly['monthly_revenue'] = df_user_monthly.apply(calculate_revenue, axis=1)

print(df_user_monthly.head(20))
print(df_user_monthly.info())
```

<div class="alert alert-block alert-success">
<b>Reviewer's comment, Iteration 1</b> <a class="tocSkip"></a>

Buen pipeline aquí: agregaste el uso por usuario-mes, integraste los datos del plan y calculaste los ingresos mensuales aplicando las reglas de redondeo requeridas (redondeo hacia arriba (ceil) de minutos por llamada y redondeo hacia arriba (ceil) de GB por mes).

## Prueba las hipótesis estadísticas

H0: No hay diferencias en los ingresos promedio de los usuarios de planes ultimate y surf.

H1: Hay diferencia en el ingreso promedio de los planes.

Alpha: 0.01

Los ingresos promedio no guardan una distribución normal, por lo cual se debe realizar una prueba sobre igualdad de las medias, con una incertidumbre del 1%

```python
# Prueba las hipótesis

# Separar los ingresos por plan
surf_revenue = df_user_monthly[df_user_monthly['plan'] == 'surf']['monthly_revenue']
ultimate_revenue = df_user_monthly[df_user_monthly['plan'] == 'ultimate']['monthly_revenue']

# Realizar prueba t para muestras independientes
alpha = 0.01  # Nivel de significancia más estricto
results = st.ttest_ind(surf_revenue, ultimate_revenue, equal_var=False)

print(f'Valor p: {results.pvalue}')
if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")

# Calcula las medias para interpretar la dirección de la diferencia
print(f"Ingreso promedio plan Surf: ${surf_revenue.mean():.2f}")
print(f"Ingreso promedio plan Ultimate: ${ultimate_revenue.mean():.2f}")
```

H0: no hay diferencia en el ingreso promedio de los usuarios del área de NY-NJ y los usuarios de otras regiones

H1: Hay diferencias en el ingreso promedio de los usuarios del área de NY-NJ y los usuarios de otras regiones.

Alpha: 0.01

Los ingresos promedio no guardan una distribución normal, por lo cual se debe realizar una prueba sobre igualdad de las medias, con una incertidumbre del 1%

```python
# Prueba las hipótesis

# Usuarios de NY-NJ del dataframe df_users_clean
df_users_clean['ny_nj'] = df_users_clean['state'].isin(['NY', 'NJ'])

# Agregar la información a df_user_monthly
df_user_monthly_region = df_user_monthly.merge(df_users_clean[['user_id', 'ny_nj']], on='user_id')

# Estadísticas descriptivas por región
regional_stats = df_user_monthly_region.groupby('ny_nj')['monthly_revenue'].agg(['mean', 'median', 'std', 'count'])
print("Estadísticas de ingresos por región:")
print(regional_stats)
print()

# Segmentar los ingresos mensuales por región
ny_nj_revenue = df_user_monthly_region[df_user_monthly_region['ny_nj'] == True]['monthly_revenue']
other_revenue = df_user_monthly_region[df_user_monthly_region['ny_nj'] == False]['monthly_revenue']

print(f"Tamaño muestra NY-NJ: {len(ny_nj_revenue)}")
print(f"Tamaño muestra otras regiones: {len(other_revenue)}")

# Verificar normalidad visualmente
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5), sharey=True)

# Definir los valores mínimo y máximo para los bins
min_val = df_user_monthly_region['monthly_revenue'].min()
max_val = df_user_monthly_region['monthly_revenue'].max()

# Usar los mismos bins para ambos gráficos
bins = np.linspace(min_val, max_val, 15)

ax1.hist(ny_nj_revenue, bins=bins, alpha=0.7, color='blue')
ax1.set_title(f'NY-NJ (n={len(ny_nj_revenue)})')
ax1.set_xlabel('Ingresos mensuales')

ax2.hist(other_revenue, bins=bins, alpha=0.7, color='orange')
ax2.set_title(f'Otras regiones (n={len(other_revenue)})')
ax2.set_xlabel('Ingresos mensuales')

plt.tight_layout()
plt.show()

# Realizar prueba t para muestras independientes
alpha = 0.01  # Mismo nivel de significancia
results = st.ttest_ind(ny_nj_revenue, other_revenue, equal_var=False)

print(f'Valor p: {results.pvalue}')
if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
    print("Existe diferencia significativa entre las regiones")
else:
    print("No podemos rechazar la hipótesis nula")
    print("No hay evidencia suficiente de diferencia entre regiones")

# Mostrar las medias para interpretar la dirección
print(f"Ingreso promedio NY-NJ: ${ny_nj_revenue.mean():.2f}")
print(f"Ingreso promedio otras regiones: ${other_revenue.mean():.2f}")
```
